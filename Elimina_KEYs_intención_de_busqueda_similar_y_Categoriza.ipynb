```python
# M√≥dulo 1
# Empezamos por subir fichero
from google.colab import files
!rm * # borramos archivos previos de la sesi√≥n
subirFichero = files.upload() # subimos fichero .csv
nombreFichero = list(subirFichero.keys())[0]
rutaDescarga=nombreFichero.replace(".csv", "Clusterizado.csv")

import chardet
#esta funci√≥n espec√≠fica se la he pedido al chatGPT ;)
def detect_encoding(filepath):
    with open(filepath, 'rb') as f:
        data = f.read()
        return chardet.detect(data)['encoding']

codificacionArchivo = detect_encoding(nombreFichero)
import pandas
p0 = pandas.read_csv(nombreFichero, encoding=codificacionArchivo)
nombrePrimeraColumna = p0.columns[0]
p0= p0[[nombrePrimeraColumna]]
if nombrePrimeraColumna != 'key':
  p0.rename(columns= {nombrePrimeraColumna:'key'}, inplace=True)
print(f"Fichero {nombreFichero} subido correctamente. Pasa al siguiente apartado m√°s abajo")
```

```python
# M√≥dulo 2
#HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH
#       -------------  PARAMETROS --------------
similitudIntencionBusqueda = 0.97 # # Para eliminar keys con misma intenci√≥n de b√∫squeda => Cuanto m√°s cerca de 1 m√°s se parecen. Depende de lo que quieras afinar... Con 0.95 eliminar√° m√°s keywords, pero puede cepillarse longtails
similitudCluster = 0.70 # Minima similitud entre keys de la MISMA CATEGORIA => 0.75 es bastante bueno pero conseguir√° menos clusters. En general obtine mejores resultados cuanto m√°s grande es el listado de keys
minElemCluster = 4 # M√≠nimo n√∫mero de keys por cluster. Cuanto mayor sea menos clusters/CATEGOR√çAS encontrar√°
evitar = {"DET",  "ADP", "PRON", "INTJ", "PUNCT",  "SYM", "SPACE"} # Tipos de palabras a evitar => "DET" art√≠culos determinados (el, la, los...). M√°s posibilidad... "ADV", "AUX", "SCONJ", "CONJ", "CCONJ", "VERB", "ADJ"
interesantes = {'NOUN', 'PROPN', 'ADJ'} # 'VERB' pensable incluir verbo # Para dar nombre a las categorias. Sustantivos, nombres propios... lematizados => pensable incluir 'NUM' si existe alg√∫n n√∫mero importante por ejemplo el a√±o
juntarClustersSimilares = 0.98 # junta clusters que se parezan en un 0.98 o m√°s
#HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH

def rotulos(rotulo): # funci√≥n mierder para mostrar r√≥tulos
  print("\n\n")
  print("----------------------------------------------------------------------------------")
  print(rotulo)
  print("----------------------------------------------------------------------------------")

from tqdm import tqdm
tqdm.pandas()

numKeys = len(p0)


import re, string
from unicodedata import normalize

# ---------------------------------------------------------------------------------
#                            MISMA INTENCI√ìN DE B√öSQUEDA
# M√©todo 1 => quitamos acentos, signos de puntuaci√≥n, may√∫sculas, ordenamos palabras alfab√©ticamente
def quitarAcento(s):
  s = re.sub(
        r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1",
        normalize( "NFD", s), 0, re.I
    )

  s = normalize( 'NFC', s)

  return s

def eliminarSignos(s):
  return re.sub('[^a-z0-9 ]', '', s)


def ordenar(texto):
  #quito algunos stops b√°sicos agilizar => el modelo de transformer los pilla pero va m√°s lento
  stop = ['a', 'para', 'de', 'la', 'que', 'el', 'en', 'y', 'los', 'del', 'se', 'las', 'un', 'una', 'su', 'al', 'lo', 'sus', 'le', 'ya', 'este', 'muy', 'tambi√©n', 'me', 'nos', 'les',  'ese', 'eso', 'ellos', 'e', 'esto', 'm√≠', 'unos', 'otro', 'otras', 'otra', 'esa', 'estos', 'nada', 'cual', 'mi', 'mis', 'te', 'ti', 'tu', 'tus', 'ellas', 'nosotras', 'vosotros', 'vosotras', 'os', 'mio', 'mia', 'mios', 'mias', 'tuyo', 'tuya', 'tuyos', 'tuyas', 'suyo', 'suya', 'suyos', 'suyas', 'nuestro', 'nuestra', 'nuestros', 'nuestras', 'vuestro', 'vuestra', 'vuestros', 'vuestras', 'esos', 'esas']

  texto = quitarAcento(texto)
  texto = eliminarSignos(texto)
  spliteado = texto.lower().split()
  spliteado = [x for x in spliteado if x !='']
  spliteado = [x.rstrip("s") for x in spliteado] # quito s de plural => vale discutible...


  spliteado = [x for x in spliteado if x not in stop]

  #spliteado = sorted(spliteado)
  #return " ".join(spliteado)
  return spliteado

rotulos("M√©todo 1 => Eliminar keys con Misma intenci√≥n")
p0['keySplit']= p0['key'].apply(ordenar)
p0['keyOrdenar'] = p0['keySplit'].apply(lambda x: " ".join(sorted(x)))
p0 = p0[~p0.duplicated(subset='keyOrdenar')]
p0['keyOrdenar']= p0['keySplit'].apply(lambda x: "".join(x))
p0 = p0[~p0.duplicated(subset='keyOrdenar')]
p0['keyOrdenar']= p0['key'].apply(lambda x: quitarAcento(eliminarSignos(x)).replace(' ', ''))
p0 = p0[~p0.duplicated(subset='keyOrdenar')]
p0 = p0.drop(columns=['keyOrdenar', 'keySplit'])

print(numKeys, "registros antes de eliminar duplicados")
print(len(p0), "registros tras eliminar duplicados")


#--------------------------------------------------------------------------------------
# M√©todo 2 => agrupamos por similitud sem√°ntica y eliminamos los que sean muy similares
rotulos("M√©todo 2 => Eliminar keys con Misma intenci√≥n")
print("Instalando transformer ü§ñ para buscar similitud")
print("")
!pip install sentence_transformers

from sentence_transformers import SentenceTransformer, util
from sklearn.cluster import AgglomerativeClustering
import numpy as np

modelo = 'hackathon-pln-es/paraphrase-spanish-distilroberta' #hiiamsid/sentence_similarity_spanish_es => otro posible modelo
model = SentenceTransformer(modelo)

# Me creo una clase como navaja suiza para puntuar similitud y para clusterizar
class encodings():
    def __init__(self, lista):
        self.lista  = [x for x in lista.copy()]
        self.lista = list(set(self.lista))
        self.embeddings = model.encode(self.lista, batch_size=256, show_progress_bar=True, convert_to_tensor=True) # pasar a vectores

    def calcularSimilitud(self, key):
      query_emb = model.encode(key, convert_to_tensor=True)
      scores = util.cos_sim(query_emb, self.embeddings)[0].tolist() # puntuo similitud // opcional dot_score

      #Combino textos y puntuaciones
      doc_score_pairs = list(zip(self.lista, [round(x,2) for x in scores]))
      doc_score_pairs = [list(x) for x in doc_score_pairs]
      doc_score_pairs = sorted(doc_score_pairs, key=lambda x: x[1], reverse=True) # ordeno por puntuaciones
      self.puntuaciones = doc_score_pairs

   #La chicha => clusterizamos
    def crearClusters(self, minimoElementos=5, thres=0.5):
      self.clusters00 = []
      clusters = util.community_detection(self.embeddings , min_community_size=minimoElementos, threshold=thres, batch_size=256)
      for keywordX, cluster in enumerate(clusters):
            clusterX = []
            for sentence_id in cluster[0:]:
                clusterX.append(self.lista[sentence_id])
            self.clusters00.append(clusterX)


agrupados = p0['key'].drop_duplicates().to_list()
#agrupados = [x for x in agrupados if x not in repetidos]
procesando = encodings(agrupados)

procesando.crearClusters(minimoElementos=2, thres=similitudIntencionBusqueda)


aBorrar = [x[1:] for x in procesando.clusters00]
aBorrar = [item for sublist in aBorrar for item in sublist]
print('')
print("keyswords similares encontradas:")
print('')
display(procesando.clusters00)
print('------------------------------------------------------------')
print(len(p0), "Antes de eliminar duplicados")
p0 = p0[~p0['key'].isin(aBorrar)]
print(len(p0), "Tras de eliminar duplicados")
print('')
rotulos("M√©todo 3 => Eliminar keys con Misma intenci√≥n")
print("Instalando Modelo de Spacy ü§ñ YIIIIHAAAAAAAAAAAAAAAAA")
!pip install -U spacy
!python -m spacy download es_dep_news_trf
import es_dep_news_trf

import spacy
spacy.require_gpu()
spacy.prefer_gpu()
nlp = es_dep_news_trf.load()

import re

# de cada frase pillamos s√≥lo el lemma de las palabras que no est√©n en evitar (sustantivos, verbos...)
def palabrasLematizadas(texto0):

  def limpia(texto):
    frases = re.sub('[^0-9a-zA-Z√Ä-√ø\u00f1\u00d1.¬ø?¬°!]', ' ', texto)
    frases = re.sub(r'\s+', ' ', frases)
    frases = frases.replace("  ", " ").strip()
    frases = frases.replace("(", "").replace(")", "")
    return frases

  texto1  = limpia(texto0)

  doc = nlp(texto1)


  soloPalabras = [quitarAcento(token.lemma_.lower())
          for token in doc
          if  not token.is_stop and not token.is_punct and token.pos_ not in evitar]
  return " ".join(soloPalabras)

print("--------------------------------------------")
print("Lematizando keys:")
p0['lemas'] = p0['key'].progress_apply(palabrasLematizadas)

repetidos = p0.groupby('lemas')['key'].apply(list).reset_index()
repetidos = repetidos[repetidos['key'].apply(len)>1]

print("keyswords con misma inteci√≥n encontradas:")
print('')
display(repetidos['key'].to_list())

p1 = p0.copy()
print('------------------------------------------------------------')
print(len(p1), "registros antes de eliminar duplicados")
p1['lemasOrdenado']= p1['lemas'].apply(ordenar)
p1['lemasOrdenado'] = p1['lemasOrdenado'].apply(lambda x: " ".join(sorted(x)))
aEliminar = (p1.duplicated(subset='lemasOrdenado') )
p1 = p1[~aEliminar]
p1.drop(columns=['lemasOrdenado'], inplace=True)
print(len(p1), "registros tras eliminar duplicados")
print('------------------------------------------------------------')
#-----------------------------------------------------------------
# M√©todo 4 => repetici√≥n del m√©todo 2 pero con lemas y tras haber quitado morraya
parametro = 'lemas'
agrupados = p1[parametro].drop_duplicates().to_list()
procesando = encodings(agrupados)

procesando.crearClusters(minimoElementos=2, thres=similitudIntencionBusqueda)# 0.965 Depende de lo que quieras afinar...
rotulos("M√©todo 4 => Eliminar keys con Misma intenci√≥n")
print("keyswords similares encontradas:")
display(procesando.clusters00)
print('------------------------------------------------------------')
aBorrar = [x[1:] for x in procesando.clusters00]
aBorrar = [item for sublist in aBorrar for item in sublist]
print(len(p1), "Antes de eliminar duplicados")
p1 = p1[~p1[parametro].isin(aBorrar)]
print(len(p1), "Tras eliminar duplicados")
print('------------------------------------------------------------')

# ----------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------

#HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH
print("HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH")
rotulos('FASE 2 CATEGORIZAMOS')
print("HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH")
agrupados = p1['key'].drop_duplicates().to_list() # opcionalmente podr√≠a clusterizar por lemas
procesando = encodings(agrupados)
procesando.crearClusters(minimoElementos=minElemCluster, thres=similitudCluster)# 0.75 optimo Depende de lo que quieras afinar...
#print(len(procesando.clusters00), 'categorias encontradas')
elementos = [item for sublist in procesando.clusters00 for item in sublist]
print(len(set(elementos)), 'elementos categorizados')

#----------------------------------------------------------------------------------
# Doy nombre a los clusters con sustantivos m√°s usados
import re
from nltk.tag import pos_tag
from collections import Counter
from nltk import ngrams

rotulos("Fase 2A Damos nombre a las categor√≠as con sustantivos m√°s usados...") #Empieza la magia
import spacy
spacy.require_gpu()
spacy.prefer_gpu()
nlp = es_dep_news_trf.load()

def palabrasMasUsadas(clusterX): #
  clusterX = [str(x) for x in clusterX]
  texto = ". ".join(clusterX).replace("..", ".").replace("!.", "!").replace("?.", "?").replace(":.", ":")
  def limpia(texto):
    frases = re.sub('[^a-zA-Z√Ä-√ø0-9\u00f1\u00d1.¬ø?¬°!]', ' ', texto)
    frases = re.sub(r'\s+', ' ', frases)
    frases = frases.replace("  ", " ").strip()
    frases = frases.replace("(", "").replace(")", "")
    return frases

  texto  = limpia(texto)

  doc = nlp(texto)

  soloPalabras = [token.lemma_.lower()
          for token in doc
          if not token.is_stop and not token.is_punct and token.pos_ in interesantes]

  from collections import Counter

  relevantes = Counter(soloPalabras)
  relevantes = relevantes.most_common()
  salida = [list(x) for x in relevantes[:3]]
  if len(salida)==0: return ''

  masAlto = salida[0][1]
  salida2 = [x[0].lower() for x in salida if x[1] > masAlto/3] #lo uso para comprobar si contiene las dos palabras del bigram m√°s usado

  salida3 = [x[0].lower() for x in salida if x[1] > len(clusterX) /2] # el lema aparece m√°s de la mitad de las veces
  if len(salida3) > 0:
    salida = salida3
  else:
    salida = [x[0].lower() for x in salida if x[1] > masAlto/1.8] # porsiaca


  if len(salida)==0: return ""

  salida = " ".join(salida)


  def devuelveNgrams(texto, n):
    xgrams = list(ngrams(texto.split(), n))
    xgrams = [" ".join(gram) for gram in xgrams]
    word_freq = Counter(xgrams)
    return word_freq


  soloPalabras0 = [token.text.lower()
        for token in doc
        if not token.is_stop and not token.is_punct and token.pos_ not in evitar]

  entero = " ".join(soloPalabras0)
  bigrams = devuelveNgrams(entero, 2).most_common()
  if len(bigrams) == 0:
    return salida
  bigram = bigrams[0]

  if len(list(set(bigram[0].split(" ")))) < 2:
    return salida


  if (len(clusterX)/bigram[1]) <2 : # si el bigram aparece en la mitad de las keys
    return bigram[0]

  palabrasBigramNoContenidas = [x for x in bigram[0].split(" ") if x not in salida2]



  if len(palabrasBigramNoContenidas) ==0 and (len(clusterX)/bigram[1]) <2 :
    return bigram[0]


  else:
    return salida

p1['cluster']=''


for lista in procesando.clusters00: # para cada cluster busco un nombre
  listaStr = [str(x) for x in lista]
  p1.loc[p1['key'].isin(lista), 'cluster']=palabrasMasUsadas(listaStr)


# -------------------------------------------------------------
rotulos('Fase2B => Bola extra: trato de clusterizar alguno m√°s usando el nombre de los clusters')
p1['cluster']= p1['cluster'].fillna('')
filtrado = p1['cluster']==''
corpus = p1[~filtrado]['cluster'].drop_duplicates().to_list() # categor√≠as
corpus_embeddings = model.encode(corpus , batch_size=256, show_progress_bar=True, convert_to_tensor=True)

queries = p1[filtrado]['lemas'].drop_duplicates().to_list() # keys que no he categorizado
queries = [x for x in queries if x !='']
queries_embeddings = model.encode(queries , batch_size=256, show_progress_bar=True, convert_to_tensor=True)

hits = util.semantic_search(queries_embeddings, corpus_embeddings, top_k=1) # encontramos la categor√≠a que m√°s se parezca

hits = [[corpus[x[0]['corpus_id']], x[0]['score']] for x in hits]

resultados =  list(zip(queries, hits))
resultados = [[x[0], x[1][0]] for x in resultados if x[1][1] > similitudCluster] # filtro con la similitud m√≠nima escogida
nuevosClusterizados = pandas.DataFrame(resultados,  columns =['lemas', 'cluster'])

#recupero categoria
p1.set_index('lemas', inplace=True)
p1.update(nuevosClusterizados.set_index('lemas'))
p1 = p1.reset_index()
print(len(nuevosClusterizados), "Recategorizados")

p1['cluster'] = p1['cluster'].str.title()
p1.drop(columns=['lemas'], inplace=True)

#--------------------------------------------------------------------
# junto clusters muy similares
rotulos("Fase 2C => Junto clusters muy similares... Si los hay...")
misQueridosClusters = p1['cluster'].drop_duplicates().to_list()
procesando = encodings(misQueridosClusters)
procesando.crearClusters(minimoElementos=2, thres=juntarClustersSimilares)
for juntos in procesando.clusters00:
  p1.loc[p1['cluster'].isin(juntos), 'cluster']=juntos[0]

# HHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHHH
# REPORTE

tablaCusters = p1.groupby('cluster')['key'].apply(list).reset_index()
tablaCusters['items']=tablaCusters['key'].apply(len)
tablaCusters.sort_values(by=['items'], ascending=False, inplace=True)
display(tablaCusters)

keysUnicas = len(p1)
numCategorias= len(p1['cluster'].drop_duplicates())
numKeysCategorizadas = len(p1[(p1['cluster']!='') & (p1['cluster'].notna())])

from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

def graficaDeQuesos(numeros, leyenda):
  pie, ax = plt.subplots(figsize=[10,6])
  plt.pie(numeros, labels = leyenda,  autopct='%.0f%%', radius=1.5)
  plt.show()

# mostramos resumen de resultados
rotulos("Gr√°fica de keys duplicadas")
print(numKeys, " Keys iniciales")
print(keysUnicas, " Keys tras eliminar con misma intenci√≥n de b√∫squeda")
graficaDeQuesos([keysUnicas,numKeys - keysUnicas], ["Keys √∫nicas", "Keys repetidas"])

rotulos("Gr√°fica de keys clusterizadas")
print(numCategorias, " N√∫mero de categorias")
print(numKeysCategorizadas, "N√∫mero de keys clusterizadas")
print(keysUnicas - numKeysCategorizadas, "N√∫mero de keys NO clusterizadas")
graficaDeQuesos([numKeysCategorizadas,keysUnicas - numKeysCategorizadas], ["Keys categorizadas", "Keys NO categorizadas"])

rotulos("Descargamos el csv clusterizado... Que tengas un dia estupendito")
p1.to_csv(rutaDescarga, index=False, encoding=codificacionArchivo)
files.download(rutaDescarga)
```
